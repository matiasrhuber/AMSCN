## Answers to "Analysis and General Questions"

##### 4.4.1: Lag Size
For the first test series, lag sizes of 0 and 1 are reasonable; for the second series, you should choose lag sizes between 0 and 3. Since the series repeats after 2 or 4 elements, the results for higher numbers don't give you any new information. The autocorrelation with lag 0 is always close to 1, the covariance is not scaled version, so if values are higher (e.g., 5 and -5 in the first time series), their covariance is also higher than 1. In addition, for the first time serives, the autocorrelation with lag 1 is close to -1, since neigbring values are opposite in sign. In the second time series, the correlations with lag 1 and 3 are negative, since the sign of the values in the list alternates at each time step. With lag 2, the correlation is positive, because the corresponding pairs conside in sign.

##### 4.4.2: Correlation Times in the DES
IAT (inter-arrival time) vs. WT (waiting time)  
Negative, however very small, correlation values indicate that in general when packets arrival in short intervals, they are more likely to wait longer.

IAT vs. ST (service time)  
Cross-correlation values for different offered traffic are all nearly zero, meaning that they are not correlated. This makes sense because IAT and ST are generated from independent random number sequences and therefore should not rely on each other.

IAT vs. SYST (system time)  
Since SYST is a combination of WT and ST, and considering IAT and ST are uncorrelated, the relation of IAT and SYST is similar to that of IAT and WT. The values are in most cases negative, indicating longer system time when packets arrive more frequently.

ST vs. SYST  
The correlation between system time and service time is high for low offered traffic values (packets don't have to wait, so the waiting time is basically 0). For higher offered traffic, the correlation becomes much lower.

WT1 vs. WT2  
For low offered traffic, the waiting times of two users are not correlated, since the packets are normally nor queued up. As the offered traffic increases, the waiting times become correlated, since they might wait in the same queue to be served by a single server.

The autocorrelation between the waiting times is high for large values of offered traffic. The interpretation is as follows: For a high offered traffic, there will be a high number of packets waiting in the queue from time to time. When the previous packet had to wait, the current packet will probably have to wait, too, since the server is slow compared to the arrival rate. On the other side, the offered traffic value is still below 1.0, thus the server is still quick enough to process all packets faster as they arrive (in the average!). Thus, the probablity is low, that a packet has to wait (at all or for a longer time) if the last packet has been served directly.

The autocorellation of IATs is close to zero for lags values different than 0. This is expected, since we generate inter-arrival times being independent from each other.

##### 4.4.2: Visualization of Covariance/Correlation
In general, the more scatter points locate in a straight line, the higher correlation between the two variables.
Regarding IAT vs. serving time, the scatter points distributions look similar for all offered traffic values, filling the lower left corner of the  coordinate system. 
This confirms the independence of IAT and serving time (as we use independent random number generators).
Regarding serving time and system time, they tend to vary in the same manner, when A is small (0.05) and there is nearly no waiting.
When rho becomes larger, they become more and more independent from each other, as the impact of waiting time becomes larger.

#### 4.4.3: Auto Correlation

The problem with N = 100 is, that in the transient phase of the simulation, the system does not behave as it does on the long run.

The correlation values are rather random, since for 100 packets, there will be no longer bursts or sparse times, when no packets arrive. Thus, the correlation between waiting times strongly depends on the random numbers and does not really show a clear behaviour.
